### Question 10 -
Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or paragraph, and return their respective count as a dictionary.
#### Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same.  

Example Output - 


```python
import nltk
from nltk import word_tokenize, pos_tag
from collections import defaultdict

def count_pos_tags(text):
    # Tokenize the text into words
    words = word_tokenize(text)

    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)

    # Initialize counters for each part of speech
    pos_counts = defaultdict(int)

    # Count the number of occurrences for each part of speech
    for word, tag in tagged_words:
        if tag.startswith('NN'):
            pos_counts['nouns'] += 1
        elif tag.startswith('VB'):
            pos_counts['verbs'] += 1
        elif tag.startswith('JJ'):
            pos_counts['adjectives'] += 1
        elif tag.startswith('PRP'):
            pos_counts['pronouns'] += 1

    return pos_counts

# Example usage 1
text = "The majestic sun slowly sank behind the rugged mountains, casting a warm golden glow across the tranquil valley. Birds chirped melodiously as they nestled in the lush green trees, while a gentle breeze whispered through the swaying leaves. A lone hiker stood mesmerized by the breathtaking vista, his heart filled with awe and gratitude for the beauty of nature."
counts = count_pos_tags(text)
print(counts)

# Example usage 2
paragraph = "I saw a beautiful sunset while walking on the beach. The waves crashed against the shore, creating a soothing sound. Seagulls flew overhead, searching for food. It was a peaceful and serene moment."
counts = count_pos_tags(paragraph)
print(counts)
```

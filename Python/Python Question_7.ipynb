Using the data from Question 4, write code to analyze the data and answer the following questions.

Insights to be drawn -                                                                    
● Get all the Earth meteorites that fell before the year 2000                             
● Get all the earth meteorites co-ordinates who fell before the year 1970                                   
● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the provided URL
url = "https://data.nasa.gov/resource/y77d-th95.json"
df = pd.read_json(url)

# Convert the year column to datetime format for easier filtering
df['year'] = pd.to_datetime(df['year'], errors='coerce')

# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = df[(df['reclat'].notnull()) & (df['year'] < '2000-01-01')]

# Plotting the Earth meteorites that fell before 2000 on a scatter plot
plt.scatter(earth_meteorites_before_2000['reclong'], earth_meteorites_before_2000['reclat'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorites that Fell Before 2000')
plt.show()

# Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_before_1970 = df[(df['reclat'].notnull()) & (df['year'] < '1970-01-01')]

# Plotting the Earth meteorites that fell before 1970 on a scatter plot
plt.scatter(earth_meteorites_before_1970['reclong'], earth_meteorites_before_1970['reclat'], alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorites that Fell Before 1970')
plt.show()

# Get all the Earth meteorites whose mass was more than 10000kg (assuming mass is in kg)
earth_meteorites_mass_gt_10000 = df[(df['mass'].notnull()) & (df['mass'] > 10000)]

# Plotting a histogram of Earth meteorites' mass distribution
plt.hist(earth_meteorites_mass_gt_10000['mass'] / 1000, bins=20)  # Converting mass from grams to kilograms
plt.xlabel('Mass (kg)')
plt.ylabel('Count')
plt.title('Distribution of Earth Meteorites Mass (> 10000kg)')
plt.show()
```

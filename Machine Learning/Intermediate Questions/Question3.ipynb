**Q-3.** Imagine you have a dataset where you have different categories of data, Now you need to find the most similar data to the given data by using any 4 different similarity algorithms. Now you have to build a model which can find the most similar data to the given data.                                                                                                                       
dataset - https://www.kaggle.com/datasets/rmisra/news-category-dataset                            
This is the Dataset You can use this dataset for this question.                   

```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.metrics import jaccard_score
from scipy.spatial.distance import euclidean
from nltk.metrics import edit_distance
```

```python
# Load the dataset
df = pd.read_json('News_Category_Dataset_v3.json', lines=True)
df.head()
```

```python
# Preprocess the text data
nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    tokens = [token for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if token not in stop_words]
    return ' '.join(tokens)

df['clean_text'] = df['headline'] + ' ' + df['short_description']
df['clean_text'] = df['clean_text'].apply(preprocess_text)

# Example input data
given_data = "Example input data"  # Replace with your own input data

# Vectorize the text data
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['clean_text'])

# Convert the given data into a vector representation
given_vector = vectorizer.transform([preprocess_text(given_data)])

# Calculate cosine similarity
cosine_similarities = cosine_similarity(tfidf_matrix, given_vector).flatten()

# Calculate Jaccard similarity
jaccard_similarities = []
given_set = set(word_tokenize(preprocess_text(given_data)))
for i, text in enumerate(df['clean_text']):
    text_set = set(word_tokenize(text))
    jaccard_similarities.append(jaccard_score(given_set, text_set))

# Calculate Euclidean distance
euclidean_distances = []
given_array = given_vector.toarray().flatten()
for i, vector in enumerate(tfidf_matrix.toarray()):
    euclidean_distances.append(euclidean(given_array, vector))

# Calculate Levenshtein distance
levenshtein_distances = []
given_text = preprocess_text(given_data)
for i, text in enumerate(df['clean_text']):
    levenshtein_distances.append(edit_distance(given_text, text))

# Find the most similar data based on each algorithm
most_similar_cosine = df.iloc[cosine_similarities.argmax()]
most_similar_jaccard = df.iloc[jaccard_similarities.index(max(jaccard_similarities))]
most_similar_euclidean = df.iloc[euclidean_distances.index(min(euclidean_distances))]
most_similar_levenshtein = df.iloc[levenshtein_distances.index(min(levenshtein_distances))]

# Print the most similar data based on each algorithm
print("Most similar data based on cosine similarity:")
print(most_similar_cosine['clean_text'])
print()

print("Most similar data based on Jaccard similarity:")
print(most_similar_jaccard['clean_text'])
print()

print("Most similar data based on Euclidean distance:")
print(most_similar_euclidean['clean_text'])
print()

print("Most similar data based on Levenshtein distance:")
print(most_similar_levenshtein['clean_text'])
```

```python
import matplotlib.pyplot as plt
from scipy import stats

# Blood pressure data
before = [130, 142, 120, 135, 148, 122, 137, 130, 142, 128, 135, 140, 132, 145, 124, 128, 136, 143, 127, 139,
          135, 131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130,
          142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142, 128, 136, 140,
          132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128,
          136, 143, 127, 139, 135, 131, 127, 130, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135,
          131, 127, 130, 142, 128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130, 142,
          128, 136, 140, 132, 145, 124, 128, 136, 143, 127, 139, 135, 131, 127, 130]

after = [120, 135, 118, 127, 140, 118, 129, 124, 137, 125, 129, 132, 125, 136, 118, 122, 130, 139, 123, 132,
         131, 126, 120, 123, 139, 122, 129, 136, 127, 140, 119, 121, 129, 137, 122, 135, 129, 124, 119, 124,
         139, 123, 131, 135, 130, 125, 121, 124, 122, 129, 131, 136, 127, 141, 118, 121, 129, 137, 123, 135,
         130, 125, 121, 124, 124, 129, 135, 127, 141, 118, 121, 129, 137, 123, 135, 130, 125, 121, 124, 118,
         121, 129, 137, 123, 135, 130, 125, 121, 124, 118, 121, 129, 137, 123, 135, 130, 125, 121, 124, 119,
         121, 129, 137, 123, 135, 130, 125, 121, 124, 119, 121, 129, 137, 123, 135, 130, 125, 121, 124, 118,
         121, 129, 137, 123, 135, 130, 125, 121, 124]

# Calculate change in blood pressure
change = [a - b for a, b in zip(after, before)]

# Visualize the data
plt.figure(figsize=(10, 4))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(change, bins=10, edgecolor='black')
plt.xlabel('Change in Blood Pressure')
plt.ylabel('Frequency')
plt.title('Histogram')

# Q-Q plot
plt.subplot(1, 2, 2)
stats.probplot(change, dist="norm", plot=plt)
plt.title('Normal Probability Plot')

plt.tight_layout()
plt.show()

# Shapiro-Wilk test for normality
_, p_value = stats.shapiro(change)
alpha = 0.05

print(f'Shapiro-Wilk test p-value: {p_value}')

if p_value > alpha:
    print('The change in blood pressure follows a normal distribution.')
else:
    print('The change in blood pressure does not follow a normal distribution.')
```
